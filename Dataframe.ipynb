```python
#Mapping external values to dataframe values in Pandas
initial_data = {'First_name': ['Ram', 'Mohan', 'Tina', 'Jeetu', 'Meera'], 
        'Last_name': ['Kumar', 'Sharma', 'Ali', 'Gandhi', 'Kumari'], 
        'Age': [42, 52, 36, 21, 23], 
        'City': ['Mumbai', 'Noida', 'Pune', 'Delhi', 'Bihar']}
  
df = pd.DataFrame(initial_data, columns = ['First_name', 'Last_name',
                                                      'Age', 'City'])
  
# Create new column using dictionary
new_data = { "Ram":"B.Com",
             "Mohan":"IAS",
             "Tina":"LLB",
             "Jeetu":"B.Tech",
             "Meera":"MBBS" }
  
# combine this new data with existing DataFrame
df["Qualification"] = df["First_name"].map(new_data)
  
df
```

```python
#Change column names and row indexes in Pandas DataFrame
df=pd.DataFrame({"Name":['Tom','Nick','John','Peter'],
                 "Age":[15,26,17,28]})
  
# Creates a dataFrame with
# 2 columns and 4 rows
df
```

```python
df.columns =['Col_1', 'Col_2']
  
# Change the row indexes
df.index = ['Row_1', 'Row_2', 'Row_3', 'Row_4']
df
```

```python
#Using rename() function with dictionary to change a single column
# let's change the first column name
# from "A" to "a" using rename() function
df = df.rename(columns = {"Col_1":"Mod_col"})
  
df
```

```python
#Change multiple column names simultaneously –
df = df.rename({"Mod_col":"Col_1","B":"Col_2"}, axis='columns')
  
df
```

```python
#Using Lambda Function to rename the columns
df = df.rename(columns=lambda x: x+'x')
  
# this will modify all the column names
df
```

```python
# Using values attribute to rename the columns
df.columns.values[1] = 'Student_Age'
  
# this will modify the name of the first column
df
```

```python
# change the row index using the Lambda function.
df = pd.DataFrame({"A":['Tom','Nick','John','Peter'],
                   "B":[25,16,27,18]})
  
# this will increase the row index value by 10 for each row
df = df.rename(index = lambda x: x + 10)
  
df
```

```python
# if we want to change the row indexes and column names simultaneously, 
# then it can be achieved using rename() function and passing both column and index attribute as the parameter.

df = df.rename(index = lambda x: x + 5,
               columns = lambda x: x +'x')
   
# increase all the row index label by value 5
# append a value 'x' at the end of each column name. 
df
```
